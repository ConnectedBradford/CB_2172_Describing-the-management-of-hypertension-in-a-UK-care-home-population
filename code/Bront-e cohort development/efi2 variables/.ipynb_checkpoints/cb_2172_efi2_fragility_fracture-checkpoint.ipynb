```r
 CREATE OR REPLACE TABLE `yhcr-prd-bradfor-bia-core.CB_2172.cb_2172_efi2_fragfracture` as (

  with a as (
  SELECT 
  person_id,
  dateevent,
  row_number() over(partition by person_id order by dateevent) code_sequence
  FROM  `yhcr-prd-bradfor-bia-core.CB_FDM_PrimaryCare.tbl_srcode`
  WHERE snomedcode in(SELECT SNOMEDCT_CONCEPTID FROM `yhcr-prd-bradfor-bia-core.ref_Lookups.lkp_EFI2_Codelist` where deficit like 'Fragility fracture%')),

  snomed as (
  SELECT 
  m.person_id,
  m.first_episodestartdate,
  a.dateevent as frag_fracture_date,
  FROM `yhcr-prd-bradfor-bia-core.CB_2172.care_home_cohort_v1` m
  LEFT JOIN a ON
  a.person_id= m.person_id AND a.dateevent < casT(m.first_episodestartdate as datetime) AND a.dateevent > date_add(m.birth_date, interval 54 year) AND code_sequence = 1),


b as (SELECT 
  person_id,
  dateevent,
  row_number() over(partition by person_id order by dateevent) code_sequence
  FROM  `yhcr-prd-bradfor-bia-core.CB_FDM_PrimaryCare.tbl_srcode`
  WHERE ctv3code in(SELECT CTV3 FROM `yhcr-prd-bradfor-bia-core.ref_Lookups.lkp_EFI2_Codelist` where deficit like 'Fragility fracture%')),

  ctv3 as (
  SELECT 
  m.person_id,
  m.first_episodestartdate,
  a.dateevent as frag_fracture_date,
  FROM `yhcr-prd-bradfor-bia-core.CB_2172.care_home_cohort_v1` m
  LEFT JOIN a ON
  a.person_id= m.person_id AND a.dateevent < casT(m.first_episodestartdate as datetime) AND a.dateevent > date_add(m.birth_date, interval 54 year)AND code_sequence = 1),

cohort as (
select
person_id, snomed.frag_fracture_date
from snomed 

union distinct

select 
person_id, ctv3.frag_fracture_date
from ctv3
order by person_id)


select
distinct person_id,
case when frag_fracture_date is not null then 'yes'
ELSE NULL
end as frag_fracture_diagnosis
from cohort)

  



```
